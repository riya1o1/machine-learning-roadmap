# Linear Algebra for Machine Learning

**What is Linear Algebra?**  
Linear algebra is the math of vectors and matrices. In ML, every dataset, every model weight, every prediction is a vector or matrix. You cannot understand how ML works without this.

**Why does it matter?**  
- A dataset of 1000 samples with 5 features = a 1000×5 matrix
- A neural network layer = a matrix multiplication
- An image = a matrix of pixel values
- Word embeddings = vectors in high-dimensional space

**What you'll learn:**
- Scalars, Vectors, Matrices, Tensors
- Vector operations — addition, dot product, magnitude
- Matrix operations — multiplication, transpose, inverse
- Special matrices
- Eigenvalues and Eigenvectors
- Where all of this shows up in real ML

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Scalars, Vectors, Matrices, Tensors

These are just different ways to organise numbers:

| Name | Shape | ML Example |
|------|-------|------------|
| Scalar | single number | learning rate = 0.01 |
| Vector | 1D array | one training sample with 5 features |
| Matrix | 2D array | dataset of 100 samples × 5 features |
| Tensor | 3D+ array | batch of 32 images, each 28×28 pixels |

```python
# Scalar — just a number
learning_rate = 0.01
print('Scalar:', learning_rate)

# Vector — one training sample (5 features)
sample = np.array([1.5, 2.3, 0.8, 4.1, 3.0])
print('Vector:', sample)
print('Shape:', sample.shape)       # (5,)

# Matrix — full dataset (100 samples, 5 features)
dataset = np.random.randn(100, 5)
print('\nMatrix shape:', dataset.shape)  # (100, 5)

# Tensor — batch of images (32 images, 28x28 pixels, 3 color channels)
images = np.random.randn(32, 28, 28, 3)
print('Tensor shape:', images.shape)    # (32, 28, 28, 3)
```

## 2. Vectors

A vector is a point in space — or a direction. In ML, each data point is a vector. The distance between vectors tells us how similar two data points are.

```python
a = np.array([2, 3])
b = np.array([1, 4])

# Vector addition — combining features
print('a + b =', a + b)

# Scalar multiplication — scaling a vector
print('2 * a =', 2 * a)

# Magnitude (length) of a vector — how far from origin
magnitude = np.linalg.norm(a)
print('|a| =', magnitude)           # sqrt(2² + 3²) = sqrt(13)

# Unit vector — vector with magnitude 1 (used in normalisation)
unit_a = a / np.linalg.norm(a)
print('Unit vector:', unit_a)
print('Magnitude of unit vector:', np.linalg.norm(unit_a))  # always 1
```

```python
# Visualise vectors
fig, ax = plt.subplots(figsize=(5, 5))

ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a = [2,3]')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b = [1,4]')
ax.quiver(0, 0, (a+b)[0], (a+b)[1], angles='xy', scale_units='xy', scale=1, color='green', label='a+b = [3,7]')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 8)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.legend()
ax.set_title('Vector Addition')
ax.grid(True)
plt.show()
```

## 3. Dot Product

The dot product is the single most important operation in ML. A neural network layer is just a dot product. Linear regression prediction is a dot product.

**Formula:** a · b = a₁b₁ + a₂b₂ + ... + aₙbₙ

**Geometric meaning:** measures how much two vectors point in the same direction.
- Dot product = 0 → vectors are perpendicular (orthogonal)
- Dot product > 0 → vectors point in similar direction
- Dot product < 0 → vectors point in opposite directions

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Dot product — manual
manual = a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
print('Manual dot product:', manual)        # 1*4 + 2*5 + 3*6 = 32

# Dot product — NumPy
print('np.dot:', np.dot(a, b))              # 32
print('@ operator:', a @ b)                 # 32 — same thing

# Real ML use — linear regression prediction
# prediction = x1*w1 + x2*w2 + x3*w3 = dot(features, weights)
features = np.array([5.0, 3.2, 1.8])       # one sample
weights  = np.array([0.4, 0.3, 0.2])       # learned weights
bias     = 0.5
prediction = np.dot(features, weights) + bias
print('\nPrediction:', prediction)
```

## 4. Cosine Similarity

Used in NLP to measure how similar two texts are. Two identical sentences = cosine similarity of 1. Completely opposite = -1.

```python
def cosine_similarity(a, b):
    # cos(θ) = (a · b) / (|a| × |b|)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Similar vectors
v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 6])        # same direction, just scaled
print('Similar vectors:', cosine_similarity(v1, v2))   # 1.0

# Different vectors
v3 = np.array([1, 0, 0])
v4 = np.array([0, 1, 0])
print('Orthogonal vectors:', cosine_similarity(v3, v4)) # 0.0

# Real ML use: word embeddings
# king = [0.8, 0.3, 0.9], queen = [0.7, 0.4, 0.8]
king  = np.array([0.8, 0.3, 0.9])
queen = np.array([0.7, 0.4, 0.8])
man   = np.array([0.2, 0.9, 0.1])
print('\nking vs queen:', cosine_similarity(king, queen).round(3))
print('king vs man:', cosine_similarity(king, man).round(3))
```

## 5. Matrix Operations

A matrix is a 2D grid of numbers. In ML, your entire dataset is a matrix. Every neural network layer transforms one matrix into another.

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Matrix multiplication — (2×3) @ (3×2) = (2×2)
# Rule: columns of A must equal rows of B
C = A @ B
print('A @ B =\n', C)
print('Shape:', C.shape)        # (2, 2)

# Transpose — flip rows and columns
print('\nA.T =\n', A.T)         # (3, 2) from (2, 3)

# Element-wise multiplication (NOT matrix multiplication)
X = np.array([[1, 2], [3, 4]])
Y = np.array([[5, 6], [7, 8]])
print('\nElement-wise X*Y =\n', X * Y)   # [1*5, 2*6], [3*7, 4*8]
```

```python
# Real ML use — forward pass through a neural network layer
# output = input @ weights + bias

batch_size   = 4    # 4 samples
input_size   = 3    # 3 input features
output_size  = 2    # 2 neurons in this layer

X       = np.random.randn(batch_size, input_size)   # (4, 3)
W       = np.random.randn(input_size, output_size)  # (3, 2)
b       = np.random.randn(output_size)              # (2,)

output  = X @ W + b     # (4, 3) @ (3, 2) + (2,) = (4, 2)
print('Input shape:', X.shape)
print('Weights shape:', W.shape)
print('Output shape:', output.shape)    # (4, 2) — 4 samples, 2 outputs each
```

## 6. Special Matrices

These appear constantly in ML derivations and implementations.

```python
# Identity matrix — multiplying by it changes nothing (like multiplying by 1)
I = np.eye(3)
print('Identity matrix:\n', I)

# Diagonal matrix — only values on the diagonal
D = np.diag([1, 2, 3])
print('\nDiagonal matrix:\n', D)

# Symmetric matrix — A = A.T (appears in covariance matrices)
S = np.array([[1, 2, 3],
              [2, 5, 4],
              [3, 4, 6]])
print('\nIs symmetric:', np.allclose(S, S.T))  # True

# Inverse — A @ A_inv = Identity (used in solving linear equations)
A = np.array([[2, 1],
              [5, 3]])
A_inv = np.linalg.inv(A)
print('\nA @ A_inv =\n', (A @ A_inv).round(10))  # identity matrix
```

## 7. Eigenvalues and Eigenvectors

**The idea:** when you multiply a matrix by a special vector (eigenvector), the vector only gets scaled — it doesn't change direction.

**Formula:** A · v = λ · v  
where v = eigenvector, λ = eigenvalue (how much it scales)

**Why ML cares:** PCA (Principal Component Analysis) — used for dimensionality reduction — is entirely built on eigenvectors. The eigenvectors of the covariance matrix tell you the directions of maximum variance in your data.

```python
A = np.array([[3, 1],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print('Eigenvalues:', eigenvalues)      # [4, 2]
print('Eigenvectors:\n', eigenvectors)  # columns are eigenvectors

# Verify: A @ v = λ @ v
v = eigenvectors[:, 0]                  # first eigenvector
lam = eigenvalues[0]                    # first eigenvalue

print('\nA @ v =', A @ v)
print('λ * v =', lam * v)
print('Equal:', np.allclose(A @ v, lam * v))  # True
```

```python
# Real ML use — PCA from scratch using eigendecomposition
np.random.seed(42)

# Generate correlated 2D data
mean = [0, 0]
cov  = [[3, 2], [2, 2]]          # covariance matrix — features are correlated
X    = np.random.multivariate_normal(mean, cov, 200)

# Step 1: compute covariance matrix
X_centered = X - X.mean(axis=0)
cov_matrix = (X_centered.T @ X_centered) / (len(X) - 1)

# Step 2: eigendecomposition of covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 3: sort by largest eigenvalue — most important direction first
idx = np.argsort(eigenvalues)[::-1]
eigenvalues  = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Step 4: project data onto first principal component
X_pca = X_centered @ eigenvectors[:, 0:1]  # reduce 2D to 1D

# Variance explained
variance_explained = eigenvalues / eigenvalues.sum()
print('Variance explained by each component:', variance_explained.round(3))

# Visualise
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.scatter(X[:, 0], X[:, 1], alpha=0.4, s=10)
for i, (val, vec) in enumerate(zip(eigenvalues, eigenvectors.T)):
    ax1.quiver(0, 0, vec[0]*val, vec[1]*val,
               angles='xy', scale_units='xy', scale=1,
               color=['red', 'blue'][i], width=0.02,
               label=f'PC{i+1} (var={variance_explained[i]:.2f})')
ax1.set_title('Original Data + Principal Components')
ax1.legend()
ax1.set_aspect('equal')

ax2.hist(X_pca, bins=30, color='green', alpha=0.7)
ax2.set_title('Data projected onto PC1')
ax2.set_xlabel('PC1')

plt.tight_layout()
plt.show()
```

## 8. Norms — Measuring Distance

Norms measure the size of a vector. In ML they're used in regularisation (L1, L2) to prevent overfitting.

```python
w = np.array([3.0, -4.0, 1.0, 2.0])

# L1 norm — sum of absolute values (used in Lasso regularisation)
l1 = np.linalg.norm(w, ord=1)
print('L1 norm:', l1)           # |3| + |-4| + |1| + |2| = 10

# L2 norm — square root of sum of squares (used in Ridge regularisation)
l2 = np.linalg.norm(w, ord=2)
print('L2 norm:', l2)           # sqrt(9 + 16 + 1 + 4) = sqrt(30)

# Distance between two vectors — how different are two data points?
a = np.array([1.0, 2.0, 3.0])
b = np.array([4.0, 6.0, 3.0])
distance = np.linalg.norm(a - b)
print('\nEuclidean distance between a and b:', distance)  # used in KNN
```

## Mini Project — Implement Linear Regression using Linear Algebra

**Task:** Solve linear regression analytically using the Normal Equation — no gradient descent, pure linear algebra.

**Normal Equation:** w = (XᵀX)⁻¹ Xᵀy

This gives the exact optimal weights in one step. Understanding this equation requires everything we just covered: transpose, matrix multiplication, and inverse.

```python
np.random.seed(42)

# Generate data: y = 3x + 2 + noise
X_raw = np.random.rand(50, 1) * 10           # 50 samples, 1 feature
y     = 3 * X_raw.flatten() + 2 + np.random.randn(50) * 1.5

# Add bias column (column of 1s) — standard trick for intercept
X = np.hstack([np.ones((50, 1)), X_raw])     # (50, 2)

print('X shape:', X.shape)   # (50, 2) — bias + feature
print('y shape:', y.shape)   # (50,)
```

```python
# Normal Equation: w = (XᵀX)⁻¹ Xᵀy
# Breaking it down step by step:

XtX     = X.T @ X                   # step 1: Xᵀ @ X  →  (2,2)
XtX_inv = np.linalg.inv(XtX)        # step 2: inverse  →  (2,2)
Xty     = X.T @ y                   # step 3: Xᵀ @ y  →  (2,)
w       = XtX_inv @ Xty             # step 4: combine  →  (2,)

print('Learned weights:')
print(f'  Intercept (bias): {w[0]:.4f}  — true value: 2')
print(f'  Slope (weight):   {w[1]:.4f}  — true value: 3')
```

```python
# Predict and evaluate
y_pred = X @ w

# Mean Squared Error
mse = np.mean((y - y_pred) ** 2)
print('MSE:', mse.round(4))

# Visualise
plt.figure(figsize=(7, 4))
plt.scatter(X_raw, y, alpha=0.6, label='Data')
plt.plot(X_raw, y_pred, color='red', linewidth=2, label=f'y = {w[1]:.2f}x + {w[0]:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression via Normal Equation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Common Mistakes to Avoid

1. **Confusing `*` and `@`** — `A * B` is element-wise, `A @ B` is matrix multiplication. In ML you almost always want `@`.
2. **Shape errors in matrix multiplication** — (m×n) @ (n×p) = (m×p). The inner dimensions must match. Print `.shape` constantly.
3. **Forgetting to transpose** — `X.T @ X` is not the same as `X @ X.T`. One gives (features×features), the other gives (samples×samples).
4. **Singular matrices** — not all matrices have an inverse. If `np.linalg.det(A) == 0`, the matrix is singular and has no inverse.
5. **Treating eigenvectors as rows** — `np.linalg.eig` returns eigenvectors as **columns**, not rows. Always use `eigenvectors[:, i]` for the i-th eigenvector.
